## 13.Bir sanatçının (örneğin, 'The Beatles) şarkılarının dans edilebilirlik ortalamasını bul

##  18. Her sanatçı için en uzun şarkının süresini bulun(30’a kadar)

```python
longest_tracks = data.loc[data.groupby('artists')['duration_ms'].idxmax()]

longest_tracks = longest_tracks[['artists', 'track_name', 'duration_ms']]
longest_tracks_sorted = longest_tracks.sort_values(by='duration_ms', ascending=False)
top_30_longest_tracks = longest_tracks_sorted.head(30)
top_30_longest_tracks
```

## 20.Bir şarkının enstrümantal olup olmadığını kontrol eden sorgu

```python
track_name = "Release Him" 
is_instrumental = data[data["track_name"] == track_name]["instrumentalness"].values

if len(is_instrumental) > 0:
    is_instrumental = is_instrumental[0] >= 0.5
else:
    is_instrumental = None
if is_instrumental is None:
    print(f"{track_name} şarkısı verilerde bulunamadı.")
elif is_instrumental:
    print(f"{track_name} şarkısı enstrümantal.")
else:
    print(f"{track_name} şarkısı enstrümantal değil.")
```
