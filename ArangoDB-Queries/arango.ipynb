
## 3- Her oyuncu kaç filmde oynadı?


```python
query = """FOR actor IN actors
    LET filmCount = (
        FOR filmActor IN film_actors
            FILTER filmActor.actor_id == actor.actor_id
            COLLECT WITH COUNT INTO count
            RETURN count
    )
    RETURN { "Actor": CONCAT(actor.first_name, " ", actor.last_name), "Film Count": filmCount[0] }

"""
cursor = db.aql.execute(query)

for doc in cursor:
    print(doc)
```

## 4- Birden fazla DVD'yi iade etmeyen kaç müşteri var?

```python
query = """FOR customer IN customers
    LET rentalCount = (
        FOR rental IN rentals
            FILTER rental.customer_id == customer.customer_id AND rental.return_date == null
            COLLECT WITH COUNT INTO count
            RETURN count
    )
    FILTER rentalCount[0] > 1
    RETURN {"ID": customer.customer_id, "CUSTOMER": CONCAT(customer.first_name, " ", customer.last_name) }


"""
cursor = db.aql.execute(query)

for doc in cursor:
    print(doc)
```

##  11- En uzun süre kirada kalmış filmleri bulun

```python
query = """FOR film IN films
    LET rental_duration = (
        FOR inventory IN inventories
            FILTER inventory.film_id == film.film_id
            FOR rental IN rentals
                FILTER rental.inventory_id == inventory.inventory_id
                RETURN DATE_DIFF(rental.rental_date, rental.return_date, "hours")
    )
    LET max_rental_duration = MAX(rental_duration)
    SORT max_rental_duration DESC
    LIMIT 5
    RETURN {
        "Film": film.title,
        "Rental Duration (Hours)": max_rental_duration
    }


"""
cursor = db.aql.execute(query)

for doc in cursor:
    print(doc)
```

## 12- En az kiralanan 5 film hangileridir?

```python
query = """FOR film IN films
    FOR inventory IN inventories
        FILTER inventory.film_id == film.film_id
        LET rentalCount = (
            FOR rental IN rentals
                FILTER rental.inventory_id == inventory.inventory_id
                COLLECT WITH COUNT INTO count
                RETURN count
        )
    LET total_rentals = FIRST(rentalCount)
    SORT total_rentals ASC
    LIMIT 5
    RETURN {
        "Film ID": inventory.film_id,
        "Film Name": film.title,
        "Rental Count": total_rentals
    }


"""
cursor = db.aql.execute(query)

for doc in cursor:
    print(doc)
```

## 13- En fazla kazanç sağlayan 5 müşteriyi bulun

```python
query = """FOR customer IN customers
    LET totalPayments = (
        FOR payment IN payments
            FILTER payment.customer_id == customer.customer_id
            RETURN payment.amount
    )
    LET totalAmount = SUM(totalPayments)
    SORT totalAmount DESC
    LIMIT 5
    RETURN {
        "Customer ID": customer.customer_id,
        "Customer Name": CONCAT(customer.first_name, " ", customer.last_name),
        "Total Revenue": totalAmount
    }


"""
cursor = db.aql.execute(query)

for doc in cursor:
    print(doc)
```

##  20- En az kazanç sağlayan müşterilerin ortalama ödeme miktarı 

```python
query = """FOR customer IN customers
    LET totalPayments = (
        FOR payment IN payments
            FILTER payment.customer_id == customer.customer_id
            RETURN payment.amount
    )
    LET totalAmount = SUM(totalPayments)
    SORT totalAmount ASC
    LIMIT 5
    LET avgPayment = AVERAGE(totalPayments)
    RETURN {
        "Customer ID": customer.customer_id,
        "Customer Name": CONCAT(customer.first_name, " ", customer.last_name),
        "Average Payment": avgPayment
    }



"""
cursor = db.aql.execute(query)

for doc in cursor:
    print(doc)
```

```python

# AQL ile customers koleksiyonunu çek
customers_query = 'FOR customer IN customers RETURN customer'
customers_cursor = db.aql.execute(customers_query)
customers_data = [doc for doc in customers_cursor]

# AQL ile payments koleksiyonunu çek
payments_query = 'FOR payment IN payments RETURN payment'
payments_cursor = db.aql.execute(payments_query)
payments_data = [doc for doc in payments_cursor]

# Pandas DataFrame'e dönüştür
customers_df = pd.DataFrame(customers_data)
payments_df = pd.DataFrame(payments_data)

# Her müşteri için toplam ve ortalama ödeme miktarını hesapla
payment_stats = payments_df.groupby('customer_id').agg(
    totalPayments=('amount', 'sum'),
    avgPayment=('amount', 'mean')
).reset_index()

# En az ödeme yapan 5 müşteriyi bul ve sıralama
sorted_payments = payment_stats.sort_values(by='totalPayments', ascending=True).head(5)

# Müşteri bilgilerini ödemelerle birleştir
result = pd.merge(sorted_payments, customers_df, on='customer_id')

# İstenen formatta sonuçları oluştur
result['Customer Name'] = result['first_name'] + ' ' + result['last_name']
result = result[['customer_id', 'Customer Name', 'avgPayment']]

# Sonuçları göster
print(result)
```
